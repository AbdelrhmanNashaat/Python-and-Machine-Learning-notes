```python
x = [1 , 2 , 4 ,5 , 6]
y = [3 , 4 , 8 , 15 , 12]
plt.title('Linear Graph' , fontdict = {'fontname' : 'Comic Sans MS' , 'fontsize' : '18'}) # title for the graph
plt.xlabel('Time') # give x axis label
plt.ylabel('Price') # give y axis label
plt.plot(x,y , color = 'blue' , label = 'y = 2x' , linewidth = 2 , marker = 'x' , linestyle = '--') # plot means draw
plt.legend() # show label iside graph like key map
plt.show()
```

```python
x = [1 , 2 , 4 ,5 , 6]
y = [3 , 4 , 8 , 15 , 12]

plt.title('Linear Graph')
plt.xlabel('Time')
plt.ylabel('Price')

plt.plot(x,y , color = 'blue' , label = 'y = 2x' , linewidth = 2 , marker = 'x' , linestyle = '--')

axis = plt.gca() # get current axis
axis.spines['top'].set_visible(False)
axis.spines['right'].set_visible(False)

plt.legend()
plt.show()
```

```python
exp_values = [1400 , 600 , 300 , 300 , 250]
exp_labels = ['Home Rent' , 'Food' , 'Phone/Internet' , 'Car' , 'Other Utilities']
plt.pie(exp_values,labels = exp_labels , autopct='%0.3f%%' , explode = [0 , 0 , 0.5 , 0 ,0] ) # make it bigger
plt.show()
```
