```python
x = np.linspace(0,100)

plt.subplot(2,2,1)
y = 3 * x
plt.plot(x,y,c='r')

plt.subplot(2,2,2)
y = -3 * x
plt.plot(x,y,c='b')

plt.subplot(2,2,3)
y = x**2
plt.plot(x,y,c='y')

plt.subplot(2,2,4)
y = x**0.5
plt.plot(x,y,c='g')
```

```python
x = np.linspace(0,100)

plt.subplot(2,2,1)
y = 3 * x
plt.plot(x,y,c='r')

plt.subplot(2,2,2)
y = -3 * x
plt.plot(x,y,c='b')

plt.subplot(2,2,3)
y = x**2
plt.plot(x,y,c='y')

plt.subplot(2,2,4)
y = x**0.5
plt.plot(x,y,c='g')


# to control spaces between plots
plt.subplots_adjust(left = 0 , right = 1.5 
                    , bottom = 0 , top = 1.5) # this to contol start point in each direction
```

```python
x = np.linspace(0,100)

plt.subplot(2,2,1)
y = 3 * x
plt.plot(x,y,c='r')

plt.subplot(2,2,2)
y = -3 * x
plt.plot(x,y,c='b')

plt.subplot(2,2,3)
y = x**2
plt.plot(x,y,c='y')

plt.subplot(2,2,4)
y = x**0.5
plt.plot(x,y,c='g')


# to control spaces between plots
plt.subplots_adjust(wspace = 0.5 , hspace = 0.2) # wspace : width space , hspace : hieght space 
```
